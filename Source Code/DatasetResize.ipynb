Getting Directory Length Function:

```python
def get_directory_length(directory):

    # List all files and directories in the given directory
    items = os.listdir(directory)

    # Return the length of the list

    return len(items)
```

Resizing Criteria Function:

```python
def resize_criteria(image):

    # Intializing the condition that the image does not meet criteria
    yes = 0

    # Read the image
    image = cv2.imread(image)

    # Check if image is successfully loaded
    if image is not None:
        # Get width and height of the image
        height, width, _ = image.shape

        # Calculate aspect ratio
        aspect_ratio = width / height

        # Check if both width and height are greater than 256
        if aspect_ratio <= 1.5:
            yes = 1

    return yes
```

```python
image_1 = cv2.imread(CNV_images[0])
image_1 = cv2.cvtColor(image_1, cv2.COLOR_BGR2RGB)

plt.figure()
plt.imshow(np.divide(image_1,np.max(np.max(image_1))),cmap='gray')
plt.title('Image 1 of CNV')
plt.show()
```

The image appears have less than a 1.5:1 aspect ratio, so resize_criteria should return a 1

```python
resize_criteria(CNV_images[0])
```

Checking to see how much the dataset can be cropped to be square images (256 x 256 pixels) based on aspect ration threshold (1.5:1):

```python
new_CNV_length = 0

for i in range(CNV_length):
  if resize_criteria(CNV_images[i]) == 1:
    new_CNV_length += 1
    print("Image "+str(i)+" has been analyzed")
```

Resizing CNV Training Data (256x256 pixel):

```python
image_size = 256
CNV_new_length = 0

for i in range(CNV_length):
  if resize_criteria(CNV_images[i]) == 1:
    img = cv2.imread(CNV_images[i])
    img = cv2.resize(img, (image_size, image_size))
    os.chdir(output_CNV_dir)
    cv2.imwrite(CNV_images[i], img)
    os.chdir(input_CNV_dir)
    CNV_new_length += 1
    print(str(CNV_new_length)+" images has been resized.")
```
